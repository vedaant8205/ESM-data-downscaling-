```python
import netCDF4
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.mpl.ticker as cticker

# Load the NetCDF file
# nc_file = r"C:\Users\HP\Downloads\20190603.A_WCYCL1950S_CMIP6_LRtunedHR.ne30_oECv3_ICG.anvil.cam.h0.0026-01.nc"
dataset = netCDF4.Dataset(nc_file, mode="r")

# Extract latitude, longitude, and variable (e.g., surface temperature 'TS')
lat = dataset.variables["lat"][:]
lon = dataset.variables["lon"][:]
var_name = "TS"  # Change this to the variable you want to plot
data = dataset.variables[var_name][:]

# Check if the variable has a time dimension and select the first time slice
if data.ndim == 3:
    data = data[0, :, :]  # Select first time step if time is included

# Close dataset
dataset.close()

# Create figure and axis with PlateCarree projection (good for lat/lon data)
fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={"projection": ccrs.PlateCarree()})

# Plot the data with a colormap
mesh = ax.pcolormesh(lon, lat, data, cmap="jet", shading="auto", transform=ccrs.PlateCarree())

# Add colorbar
cbar = plt.colorbar(mesh, ax=ax, orientation="vertical", label=f"{var_name} values")

# Add land features
ax.add_feature(cfeature.COASTLINE, linewidth=1)
ax.add_feature(cfeature.BORDERS, linestyle="--", linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor="none", edgecolor="black", linewidth=0.5)

# Add gridlines with labels
gl = ax.gridlines(draw_labels=True, linestyle="--", linewidth=0.5, color="gray")
gl.top_labels = True  # Hide top labels
gl.right_labels = False  # Hide right labels
gl.xlocator = cticker.MaxNLocator(5)  # Adjust longitude label frequency
gl.ylocator = cticker.MaxNLocator(5)  # Adjust latitude label frequency
gl.xlabel_style = {"size": 10, "color": "black"}
gl.ylabel_style = {"size": 10, "color": "black"}

# Set title
ax.set_title(f"Spatial Distribution of {var_name} with Latitude & Longitude Grid")
ts_mapped = os.path.join(base_address, folder_name, "TS_mapped.jpg")
plt.savefig(ts_mapped)
# Show the plot
plt.show()
```
